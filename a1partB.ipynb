```python
#find link has http and judge its valid
import re
def valid(link):
    if link is not None:
        if re.search(r'http',link) is not None:
            status=requests.get(link).status_code
            if status == 200:
                return '1'
            else:
                return '0'
        else:
            return '0'
    else:
        return '0'
    
    

```

```python
for link in soup.find_all("a"):
    linkk= link.get('href')
    print(linkk)
    va= valid(linkk)
    print(va)
        
```

```python
from datetime import datetime
for link in soup.find_all("a"):
    linkk= link.get('href')
    csv=open("links.csv","a")
    va= valid(linkk)
    time = str(datetime.now())
    if va =="1":
        csv.write(link.text +link.get('href')+"_"+va+"_" +time)
        csv.close()
```
