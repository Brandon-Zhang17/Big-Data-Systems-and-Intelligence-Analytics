```python
#confirm outliers are replaced
df3[df3['bite_Year']>2018]
```

```python
#transform some variables into right type
df3['victim_zip']=df3['victim_zip'].astype(np.int64)
df3['bite_Year'] =df3['bite_Year'].astype(np.int64)
df3['bite_Month']=df3['bite_Month'].astype(np.int64)
df3['vaccination_yrs']=df3['vaccination_yrs'].astype(np.int64) 
```
