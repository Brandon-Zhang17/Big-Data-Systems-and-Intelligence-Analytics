```python
#creat the dummy variation
#why I choose this because it's appropriate
dummy_ranks = pd.get_dummies(df['bite_Month'], prefix='bite_Month')  
```

```python
dummy_ranks.head()
```

```python
#besides bite_Year, vaccination_yrs and victim_zip, add 12-1=11 dummy variations
keep = ['Gender', 'bite_Year', 'vaccination_yrs','victim_zip']
data = df[keep].join(dummy_ranks.ix[:, 'bite_Month_2':])
data.head()
data['intercept'] = 1.0
```

```python
data.head()
```

```python
#train
train_cols = data.columns[1:]
```

```python
logit = sm.Logit(data['Gender'], data[train_cols]) 
```

```python
#fitting
re = logit.fit()
```

```python
import pandas as pd
import statsmodels.api as sm
import pylab as pl
import numpy as np
```

```python
#copy train as predict
import copy  
combos = copy.deepcopy(data)  
#the same colums  
predict_cols = combos.columns[1:]  
combos['intercept'] = 1.0    
# score is the last colum
combos['predict'] = re.predict(combos[predict_cols])   
total = 0  
hit = 0  
for value in combos.values:  
  predict = value[-1]   
  Gender = int(value[0])  
   
  # if possibility>0.5, can be used  
  if predict > 0.5:  
    total += 1    
    if Gender == 1:  
      hit += 1  
```

```python
print ('Total: %d, Hit: %d, Precision: %.2f' % (total, hit, 100.0*hit/total) )
```

```python
print(re.summary())
```

```python
re.summary()
```

```python
#Confidence interval
print(re.conf_int())
```

```python
#odds ratio
print (np.exp(re.params))
```

```python
# odds ratios and 95% CI
params = re.params
conf = re.conf_int()
conf['OR'] = params
conf.columns = ['2.5%', '97.5%', 'OddRatio']
np.exp(conf)
```
